# extracting stimuli from word file

```python
with open('../ModAdjExperiment.txt', 'r') as f:
    content = f.read()
```

```python
import re

rgx = r'\[(.+?)\]\n(.+?)\n'
experiment = dict(re.findall(rgx, content))
```

```python
experiment
```

```python
len(experiment)
```

```python
import pandas as pd

df = pd.DataFrame({'tag': experiment.keys(), 'SENTENCE': experiment.values()})
df['TYPE'] = df.tag.apply(lambda x: x.split('-')[0] if x.split('-')[0] == 'filler' else 'item')
df['CONDITION'] = df.tag.apply(lambda x: x.split('-')[0] if x.split('-')[0] != 'filler' else '')
df['TENSE'] = df.tag.apply(lambda x: x.split('-')[1])
df['INDEX'] = df.tag.apply(lambda x: chr(int(x.split('-')[2]) + 64))
```

```python
items = df.loc[df.TYPE == 'item']
fillers = df.loc[df.TYPE == 'filler']
```

```python
items
```

```python
items['LIST'] = (items.index + (items.index // 4)) % 4 + 1
```

```python
for i in range(8):
    print(items.loc[4*i:4 * (i+1)-1, 'LIST'].values)
```

```python
items.loc[items.LIST == 3]
```

```python
items.LIST.unique()
```

```python
fillers['LIST'] = (fillers.index + (fillers.index // 2)) % 2 + 1
```

```python
fillers2 = fillers.copy()
fillers2['LIST'] = fillers2['LIST'] + 2
fillers_total = pd.concat((fillers, fillers2))
```

```python
fillers_total.loc[fillers_total.LIST == 2]
```

```python
experiment_total = pd.concat((items, fillers_total)).drop(columns=['tag', 'INDEX'])
```

```python
experiment_total['FILLER_TYPE'] = experiment_total['TENSE'].apply(lambda x: x if x in ['good', 'bad'] else '')
experiment_total['TENSE'] = experiment_total['TENSE'].apply(lambda x: x if x in ['PST', 'NPST'] else '')
```

```python
experiment_total['ITEM'] = df['tag']
```

```python
experiment_total.to_csv('pcibex_files/chunk_includes/experiment.csv', index=False)
```
